# Intro to Recommender Systems Lab

Complete the exercises below to solidify your knowledge and understanding of recommender systems.

For this lab, we are going to be putting together a user similarity based recommender system in a step-by-step fashion. Our data set contains customer grocery purchases, and we will use similar purchase behavior to inform our recommender system. Our recommender system will generate 5 recommendations for each customer based on the purchases they have made.

```python
import pandas as pd
import numpy as np
from scipy.spatial.distance import pdist, squareform
```

```python
data = pd.read_csv('../data/customer_product_sales.csv')
```

```python
data.head()
```

## Step 1: Create a data frame that contains the total quantity of each product purchased by each customer.

You will need to group by CustomerID and ProductName and then sum the Quantity field.

```python
df = data.groupby(['CustomerID', 'ProductName'], as_index=False).sum('Quantity')
df.head()
```

## Step 2: Use the `pivot_table` method to create a product by customer matrix.

The rows of the matrix should represent the products, the columns should represent the customers, and the values should be the quantities of each product purchased by each customer. You will also need to replace nulls with zeros, which you can do using the `fillna` method.

```python
pivot_table = pd.pivot_table(
    data=data,
    values='Quantity',
    index='ProductName',
    columns='CustomerID',
    aggfunc= 'sum',
    fill_value=0
)

pivot_table.head()
```

## Step 3: Create a customer similarity matrix using `squareform` and `pdist`. For the distance metric, choose "euclidean."

```python
pivot_table = pivot_table.T

dist_eucl = 1 / (1 + squareform(pdist(pivot_table, metric='euclidean')))
dist_eucl
```

## Step 4: Check your results by generating a list of the top 5 most similar customers for a specific CustomerID.

```python
distances = pd.DataFrame(dist_eucl, index=pivot_table.index, columns=pivot_table.index)
distances
```

```python
def top_cust_similar(distances, ID):
    distance_sort = distances.sort_values(ID, ascending=False, axis=0).loc[:, ID]
    return distance_sort[1:6].index

customer = 98159
similar_custumers = top_cust_similar(distances, customer)
print(similar_custumers.values)
```

## Step 5: From the data frame you created in Step 1, select the records for the list of similar CustomerIDs you obtained in Step 4.

```python
custumer_product_simi = pd.DataFrame()
for i in list(similar_custumers):
    cust_product = df[df['CustomerID'] == i]
    custumer_product_simi = pd.concat([custumer_product_simi, cust_product], axis=0)[['ProductName', 'Quantity']]
    
custumer_product_simi
```

## Step 6: Aggregate those customer purchase records by ProductName, sum the Quantity field, and then rank them in descending order by quantity.

This will give you the total number of each product purchased by the 5 most similar customers to the customer you selected in order from most purchased to least.

```python
product_similar = custumer_product_simi.groupby('ProductName').sum('Quantity').sort_values('Quantity', ascending=False)
product_similar
```

## Step 7: Filter the list for products that the chosen customer has not yet purchased and then recommend the top 5 products with the highest quantities that are left.

- Merge the ranked products data frame with the customer product matrix on the ProductName field.
- Filter for records where the chosen customer has not purchased the product.
- Show the top 5 results.

```python
customer_product = df[df['CustomerID'] == customer][['ProductName', 'Quantity']]
merge_table = pd.merge(left=product_similar, right=customer_product, on='ProductName', how='outer').fillna(0)

merge_table
```

```python
filter_products = merge_table[(merge_table['Quantity_x'] > 0) & (merge_table['Quantity_y'] == 0)]
filter_products['ProductName'].head(5).to_frame()
```

## Step 8: Now that we have generated product recommendations for a single user, put the pieces together and iterate over a list of all CustomerIDs.

- Create an empty dictionary that will hold the recommendations for all customers.
- Create a list of unique CustomerIDs to iterate over.
- Iterate over the customer list performing steps 4 through 7 for each and appending the results of each iteration to the dictionary you created.

```python
def top_cust_similar(distances, ID):
    distance_sort = distances.sort_values(ID, ascending=False, axis=0).loc[:, ID]
    return distance_sort[1:6].index


recomendations = dict()
customers_id = data.sort_values('CustomerID')['CustomerID'].unique()

for i in customers_id:
    similar_custumers = top_cust_similar(distances, i)
    
    custumer_product_simi = pd.DataFrame()
    for j in list(similar_custumers):
        cust_product = df[df['CustomerID'] == j]
        custumer_product_simi = pd.concat([custumer_product_simi, cust_product], axis=0)[['ProductName', 'Quantity']]
        
    product_similar = custumer_product_simi.groupby('ProductName').sum('Quantity').sort_values('Quantity', ascending=False)
    
    customer_product = df[df['CustomerID'] == customer][['ProductName', 'Quantity']]
    merge_table = pd.merge(left=product_similar, right=customer_product, on='ProductName', how='outer').fillna(0)
    
    filter_products = merge_table[(merge_table['Quantity_x'] > 0) & (merge_table['Quantity_y'] == 0)].head(5)
    recomendations.update({i: filter_products['ProductName'].values})
    
recomendations
```

##  Step 9: Store the results in a Pandas data frame. The data frame should a column for Customer ID and then a column for each of the 5 product recommendations for each customer.

```python
reco = pd.DataFrame.from_dict(recomendations)
reco
```

## Step 10: Change the distance metric used in Step 3 to something other than euclidean (correlation, cityblock, consine, jaccard, etc.). Regenerate the recommendations for all customers and note the differences.

```python
def top_cust_similar(distances, ID):
    distance_sort = distances.sort_values(ID, ascending=False, axis=0).loc[:, ID]
    return distance_sort[1:6].index

dist_cityblock = 1 / (1 + squareform(pdist(pivot_table, metric='cityblock')))
distances = pd.DataFrame(dist_cityblock, index=pivot_table.index, columns=pivot_table.index)
customers_id = data.sort_values('CustomerID')['CustomerID'].unique()

recomendations = dict()
for i in customers_id:
    similar_custumers = top_cust_similar(distances, i)
    
    custumer_product_simi = pd.DataFrame()
    for j in list(similar_custumers):
        cust_product = df[df['CustomerID'] == j]
        custumer_product_simi = pd.concat([custumer_product_simi, cust_product], axis=0)[['ProductName', 'Quantity']]
        
    product_similar = custumer_product_simi.groupby('ProductName').sum('Quantity').sort_values('Quantity', ascending=False)
    
    customer_product = df[df['CustomerID'] == customer][['ProductName', 'Quantity']]
    merge_table = pd.merge(left=product_similar, right=customer_product, on='ProductName', how='outer').fillna(0)
    
    filter_products = merge_table[(merge_table['Quantity_x'] > 0) & (merge_table['Quantity_y'] == 0)].head(5)
    recomendations.update({i: filter_products['ProductName'].values})
    
reco = pd.DataFrame.from_dict(recomendations)
reco
```
